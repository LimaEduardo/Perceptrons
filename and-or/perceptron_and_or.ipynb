# Perceptron para predizer AND e OR para saidas binárias e bipolares

Importações necessárias para rodar a aplicação

```python
import numpy as np
```

Classe Perceptron

```python
class Perceptron(object):
    def __init__(self, input_size, lr=1, epochs=100, isBipolarActivation = False):
        self.W = np.zeros(input_size+1)
        self.epochs = epochs
        self.lr = lr
        self.isBipolarActivation = isBipolarActivation
    
    def activation_fn(self, x):
        if self.isBipolarActivation:
            return 1 if x >= 0 else -1
        return 1 if x >= 0 else 0
        
 
    def predict(self, x):
        z = self.W.T.dot(x)
        a = self.activation_fn(z)
        return a
 
    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                x = np.insert(X[i], 0, 1)
                y = self.predict(x)
                e = d[i] - y
                self.W = self.W + self.lr * e * x
```
